# Kronecker Graph Models

Generating and Visualizing kronecker graph models using python

[Video Explanation](https://www.youtube.com/watch?v=Xnpt8US31cQ&ab_channel=stanfordonline)

---


```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def show_graph_with_labels(adjacency_matrix, markersize = 4, labels = None):
    if not labels:
      labels = {i: i for i in range(adjacency_matrix.shape[0])}
    rows, cols = np.where(adjacency_matrix == 1)
    edges = zip(rows.tolist(), cols.tolist())
    gr = nx.Graph()
    gr.add_edges_from(edges)
    nx.draw(gr, node_size=500, labels=labels, with_labels=True)
    plt.show()
    plt.spy(adjacency_matrix, markersize=markersize)
    plt.show()
```

```python
start_graph = np.array([[1, 1, 0], [1, 1, 1], [0, 1, 1]])
```

```python
show_graph_with_labels(start_graph, 4)
```

```python
def gen_kronecker_graph(start_graph, n_itr):
  end_graph = start_graph.copy()
  for i in range(n_itr):
    end_graph = np.kron(start_graph, end_graph)
  return
```

```python
k_graph = gen_kronecker_graph(start_graph, 2)
```

```python
show_graph_with_labels(k_graph)
```
