# Kronecker Graph Models

Generating and Visualizing kronecker graph models using python

[Kronecker Graphs](https://www.youtube.com/watch?v=Xnpt8US31cQ&ab_channel=stanfordonline)

---


```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def show_graph_with_labels(adjacency_matrix: np.ndarray, markersize: int = 4):
    g = nx.from_numpy_array(adjacency_matrix)
    nx.draw(g, node_size=500, with_labels=True)
    plt.show()
    plt.spy(adjacency_matrix, markersize=markersize)
    plt.show()
```

```python
start_graph = np.array([[1, 1, 0], [1, 1, 1], [0, 1, 1]])
```

```python
show_graph_with_labels(start_graph, 4)
```

```python
def gen_kronecker_graph(start_graph: np.ndarray, n_itr: int = 1):
  end_graph = start_graph.copy()
  for i in range(n_itr):
    end_graph = np.kron(start_graph, end_graph)
  return end_graph
```

```python
k_graph = gen_kronecker_graph(start_graph, 2)
```

```python
show_graph_with_labels(k_graph)
```
